```python
import folium
import pandas as pd
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("공공자전거 대여소 정보.csv",encoding = 'UTF-8')
```

```python
df
```

```python
df.info()
```

# 1. 종로구

```python
jonglo_df = df[df['자치구'] == '종로구'].reset_index(drop=True)
jonglo_df
```

```python
jonglo_df['위도'].mean()
```

```python
jonglo_df['경도'].mean()
```

```python
latitude = 37.57812532
longitude = 126.987327799

jonglo_map = folium.Map(location=[latitude, longitude],
               zoom_start=14, 
               width=750, 
               height=500
              )
jonglo_map
```

```python
for en in range(len(jonglo_df['경도'])):
    folium.CircleMarker([jonglo_df['위도'][en], jonglo_df['경도'][en]],
                        fill = True,
                        color = 'black',
                        radius = 2,
                        fill_color = 'black',
                        popup = jonglo_df['보관소(대여소)명'][en]).add_to( jonglo_map )

jonglo_map
```

```python
jonglo_name = jonglo_df[['보관소(대여소)명']]
jonglo_name
```

```python
jonglo_clust = jonglo_df[['위도','경도']]
jonglo_clust
```

```python
clust_model = KMeans(n_clusters = 12, n_init = 100, max_iter = 100, random_state= 34)
clust_model.fit(jonglo_clust)
centers = clust_model.cluster_centers_
pred = clust_model.predict(jonglo_clust)
a=jonglo_clust.columns[0:]
jonglo_clust[a]=jonglo_clust[a].astype(float)
```

```python
jonglo_clust['clust'] = pred
jonglo_clust['name'] = jonglo_name
jonglo_clust
```

```python
jonglo_clust['clust'].value_counts()
```

```python
jonglo_point = jonglo_clust.groupby('clust')['경도','위도'].mean().reset_index()
jonglo_point
```

```python
for en in range(len(jonglo_point['경도'])):
    folium.CircleMarker([jonglo_point['위도'][en], jonglo_point['경도'][en]],
                        fill = True,
                        color = 'red',
                        radius = 2,
                        fill_color = 'red').add_to( jonglo_map )

jonglo_map
```

```python
jonglo_list = ['평창동주민센터 앞', '신영동삼거리(북악터널방향)','독립문역 3-1번출구','서울특별시 교육청','경복궁역 7번출구 앞',
               '광화문역 7번출구 앞','종로2가 버스정류장 (종각방향)','삼청동 골목','종묘','성균관대입구사거리','동묘앞역 1번출구 뒤','동대문 종합시장 버스정류장']
```

```python
latitude = 37.57812532
longitude = 126.987327799

jonglo_final = folium.Map(location=[latitude, longitude],
               zoom_start=14,
               width=750, 
               height=500
              )
jonglo_final
```

```python
for i in jonglo_list:
    folium.CircleMarker([jonglo_df[jonglo_df['보관소(대여소)명'] == i]['위도'].item(), jonglo_df[jonglo_df['보관소(대여소)명'] == i]['경도'].item()],
                        fill = True,
                        color = 'blue',
                        radius = 3,
                        fill_color = 'blue',
                        popup = i).add_to( jonglo_final )

jonglo_final
```

```python
df1 = jonglo_df[jonglo_df['보관소(대여소)명'] == '평창동주민센터 앞']
for i in jonglo_list[1:]:
    df1 = pd.concat([df1,jonglo_df[jonglo_df['보관소(대여소)명'] == i]])

df1
```
